```python
from causalis.data.dgps import obs_linear_26_dataset

# generating synthetic observational data with a linear effect
df = obs_linear_26_dataset(return_causal_data = False)
```

```python
from causalis.data import CausalData

# create a CausalData object
causal_data = CausalData(
    df=df,
    treatment="d",
    outcome="y",
    confounders=["tenure_months",
                 "avg_sessions_week",
                 "spend_last_month",
                 "premium_user",
                 "premium_user",
                 "urban_resident"]
)

# Let's take a look at the data
causal_data.df.head()
```

```python
from causalis.inference.ate import dml_ate

# Estimate Average Treatment Effect with Double Machine Learning (ATE)
ate_result_causalis = dml_ate(causal_data, n_folds=4, store_diagnostic_data=False, random_state=123)
```

```python
print("Real ATE = 0.8")
print(f"Estimated with Causalis = {ate_result_causalis.get('coefficient')} in {ate_result_causalis.get('confidence_interval')}")
```
