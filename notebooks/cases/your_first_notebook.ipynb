```python
from causalis.dgp import obs_linear_26_dataset

# generating synthetic observational data_contracts with a linear effect
df = obs_linear_26_dataset(return_causal_data = False)
```

```python
from causalis.data_contracts import CausalData

# create a CausalData object
causaldata = CausalData(
    df=df,
    treatment="d",
    outcome="y",
    confounders=["tenure_months",
                 "avg_sessions_week",
                 "spend_last_month",
                 "premium_user",
                 "premium_user",
                 "urban_resident"]
)

# Let's take a look at the data_contracts
causaldata.df.head()
```

```python
from causalis.scenarios.unconfoundedness import IRM

model = IRM().fit(causaldata)

# Estimate Average Treatment Effect with Double Machine Learning (ATE)
result = model.estimate(score='ATE')
result.summary()
```
