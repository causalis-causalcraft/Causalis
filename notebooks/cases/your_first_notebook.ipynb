```python
from causalis.scenarios.unconfoundedness.dgp import generate_obs_hte_26

# generating synthetic observational data with a linear effect
df = generate_obs_hte_26(return_causal_data = False)
```

```python
print(f"Ground truth ATE is {df['cate'].mean()}")
```

```python
from causalis.data_contracts import CausalData

# create a CausalData object
causaldata = CausalData(
    df=df,
    treatment="d",
    outcome="y",
    confounders=["tenure_months",
                 "avg_sessions_week",
                 "spend_last_month",
                 "premium_user",
                 "premium_user",
                 "urban_resident"]
)

# Let's take a look at the data
causaldata.df.head()
```

```python
from causalis.scenarios.unconfoundedness import IRM

model = IRM().fit(causaldata)

# Estimate Average Treatment Effect with Double Machine Learning (ATE)
result = model.estimate(score='ATE')
result.summary()
```
